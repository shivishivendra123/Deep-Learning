```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
def build_model(hp):
    model = Sequential()
    counter = 0
    for i in range(hp.Int('num_layers',min_value=1,max_value=200,step=1)):
        if counter == 0:
            model.add(Dense((hp.Int('units'+str(i),min_value=8,max_value=128,step=2)),activation=hp.Choice('activation'+str(i),values=['relu','sigmoid','tanh']),input_dim=8))
        else:
            model.add(Dense((hp.Int('units'+str(i),min_value=8,max_value=128,step=2)),activation=hp.Choice('activation'+str(i),values=['relu','sigmoid','tanh'])))
        counter = counter+1

    model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer = hp.Choice('optimizer',values=['rmsprop','sgd','adam','nadam','adamdelta']),metrics=['accuracy'])
    return model
```

```python
tuner1 = keras_tuner.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5)
```

```python
tuner1.search(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
best_model = tuner1.get_best_models()[0]
best_model.summary()
```

```python
best_model.fit(X_train,y_train,epochs=2000,validation_data=(X_test,y_test),verbose=True)
```
