```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/shivendragupta/Desktop/Deep Learning/ANN/Admission_Predict.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:7],df.iloc[:,-1],test_size=.2,random_state=32)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

X_train_trs = mms.fit_transform(X_train)
X_test_trs = mms.transform(X_test)
```

```python
X_train_trs.shape
```

```python
model = Sequential()

model.add(Dense(7,activation='relu',input_dim=7))
model.add(Dense(5,activation='relu'))
model.add(Dense(3,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer='Adam',metrics=['r2_score'])
```

```python
model.fit(X_train_trs,y_train,epochs=1000)
```

```python
y_pred = model.predict(X_test_trs)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
