```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/shivendragupta/Desktop/Deep Learning/ANN/ushape.csv',header=None)
```

```python
df.head()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal')) #he_normal for relu , glorat_normal for tanh
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=1000,verbose=True,validation_split=.2)
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
