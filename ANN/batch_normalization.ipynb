```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/shivendragupta/Desktop/Deep Learning/ANN/concertriccir2.csv',header=None)
```

```python
df.head()
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c = y)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization
```

```python
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,batch_size=4,epochs=200,validation_split=.2,verbose=True)
```

```python
model_bn = Sequential()

model_bn.add(Dense(16,activation='relu',input_dim=2))
model_bn.add(BatchNormalization())
model_bn.add(Dense(16,activation='relu'))
model_bn.add(BatchNormalization())
model_bn.add(Dense(16,activation='relu'))
model_bn.add(BatchNormalization())
model_bn.add(Dense(16,activation='relu'))
model_bn.add(BatchNormalization())
model_bn.add(Dense(1,activation='sigmoid'))

model_bn.summary()
```

```python
model_bn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model_bn.fit(X,y,batch_size=4,epochs=1200,validation_split=.2,verbose=True)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X,y.astype(np.int_),clf=model_bn)
```
